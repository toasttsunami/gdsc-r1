Resolution column is of type object with the entries as string, so i extract the horizontal and vertical resolution as integers and split them into two columns

```python
def extract_resolution(string):
    # Regular expression pattern to extract numeric values
    pattern = r'(\d+)\D+(\d+)'
    match = re.search(pattern, string)
    if match:
        return int(match.group(1)), int(match.group(2))
    else:
        return None, None

# Apply the extraction function to Resolution column
df[['Horizontal Resolution', 'Vertical Resolution']] = df['Resolution'].apply(extract_resolution).apply(pd.Series)

print (df[['Vertical Resolution', 'Horizontal Resolution']])
print (df['Vertical Resolution'].unique())
print (df['Horizontal Resolution'].unique())
```

Internal Storage and RAM are stored as strings, but with just numeric values, so we can just typecast them into float, and then convert the gb values into mb

```python
df['Internal Storage'] = df['Internal Storage'].apply(lambda x: float(x.replace(' GB', '')) * 1024 if 'GB' in x else float(x.replace(' MB', '')))

df['Internal Storage'].head()
```

these columns have binary Yes No values in them so i can convert them into boolean with true if yes else false

```python
df['RAM'] = df['RAM'].apply(lambda x: float(x.replace(' GB', '')) * 1024 if 'GB' in x else float(x.replace(' MB', '')))
```

```python
binary_columns = ['Touchscreen', 'Primary Camera Available', 'Secondary Camera Available', 'Bluetooth Support', 'Wi-Fi', 'GPS Support']

for i in binary_columns:
    df[i] = df[i].apply(lambda x: bool(True) if x == 'Yes' else bool(False))
```

```python
df['Secondary MP'] = df['Secondary Camera'].apply(lambda x: int(re.findall(r'\d+MP', str(x))[0].strip('MP')) if re.findall(r'\d+MP', str(x)) else 0)
```

Deviding df_final into input and output variables

```python
features = df_final.drop('Price', axis=1)
target = df_final['Price'] 
```

Splitting the features and target data into X and y training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=55)
```

Using linear regression model from sklearn, and calculating the evaluation metrics to see how accurate the trained model is on the test dataset

```python
model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
plt.scatter(y_test, y_pred, label='Actual')
min_value = min(min(y_test), min(y_pred))
max_value = max(max(y_test), max(y_pred))

plt.plot([min_value, max_value], [min_value, max_value], color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

scaling the input variables to make regression model more accurate

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
plt.scatter(y_test, y_pred, label='Actual')
min_value = min(min(y_test), min(y_pred))
max_value = max(max(y_test), max(y_pred))

plt.plot([min_value, max_value], [min_value, max_value], color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

Using random forest regressor model and calculating its accuracy

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
plt.scatter(y_test, y_pred, label='Actual')
min_value = min(min(y_test), min(y_pred))
max_value = max(max(y_test), max(y_pred))

plt.plot([min_value, max_value], [min_value, max_value], color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
